```python
import pandas as pd
import os

file_path = r"X:\gpo_roster_files\JUNE 2024 - ROSTERS"



```

```python
medassets = os.path.join(file_path, "MedAssets_Membership_2024_06.csv")

medassets_df = pd.read_csv(medassets)
```

```python
medassets_df = medassets_df[
    ["LIC", "HIN", "Member Name", "Address1", "Address2", "City", "State", "Zip Code"]
]

medassets_df.fillna("", inplace=True)
medassets_df["HIN"] = medassets_df["HIN"].astype(str).str.lstrip('="').str.rstrip('"')
medassets_df["Zip Code"] = medassets_df["Zip Code"].astype(str).str.lstrip('="').str.rstrip('"')

medassets_df["origin"] = "MedAssets"

medassets_df.head()
```

```python
premier = os.path.join(file_path, "Premier_Membership_2024_06.txt")

premier_df = pd.read_csv(premier, sep="\t", encoding="latin1")
```

```python
premier_df = premier_df[
    [
        "GPO ID",
        "GLN",
        "Health Industry Number (HIN)",
        "Name 1",
        "Name 2",
        "Address Type",
        "Address 1",
        "Address 2",
        "Address 3",
        "City",
        "State/Province",
        "Postal Code"
    ]
]

premier_df.fillna("", inplace=True)
premier_df["GLN"] = premier_df["GLN"].astype(str).str.rstrip(".0")

premier_df["origin"] = "Premier"

premier_df.head()
```

```python
premier_df[premier_df["Health Industry Number (HIN)"] == ""]
```

```python
# compare the two dataframes and save duplicate HINs to a dict with the HIN as the key and the value as a list of the two rows
duplicates = {}

for index, row in medassets_df.iterrows():
    hin = row["HIN"]
    if hin != "":
        if hin in premier_df["Health Industry Number (HIN)"].values:
            if hin in duplicates:
                duplicates[hin].append(row)
            else:
                duplicates[hin] = [row]

for index, row in premier_df.iterrows():
    hin = row["Health Industry Number (HIN)"]
    if hin != "":
        if hin in medassets_df["HIN"].values:
            if hin in duplicates:
                duplicates[hin].append(row)
            else:
                duplicates[hin] = [row]

# create a new dataframe with the duplicates
duplicates_df = pd.DataFrame()

for key, value in duplicates.items():
    for row in value:
        duplicates_df = duplicates_df.append(row)


duplicates_df.to_csv("duplicates.csv", index=False)

```
